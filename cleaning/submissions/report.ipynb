```python
import pandas as pd
```

```python
submissions = pd.read_csv('Submission.csv')
submissions.head()
```

```python
users = pd.read_csv("users.csv")
teams = pd.read_csv("Teams.csv")

report = pd.DataFrame() #submissions[['teamId', 'score', 'userId']]

# report = report.apply(lambda row: )
```

```python
team_name_mapping = dict(zip(teams['id'], teams['name']))
report['teamName'] = submissions['teamId'].map(team_name_mapping)

# Map user names using 'id' in users.csv
user_name_mapping = dict(zip(users['id'], users['name']))
report['userName'] = submissions['userId'].map(user_name_mapping)

# adding scores
report['Scores'] = submissions['score']

# Display the updated DataFrame
print(submissions)

# Save to a new CSV file (optional)
report.to_csv('updated_submissions.csv', index=False)
```
