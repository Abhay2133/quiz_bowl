```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('Participants.csv')
df.head()
```

```python
teams = pd.DataFrame({'name':df['Team Name'], 'updatedAt': datetime.now().isoformat()})
teams.head()
```

```python
user1 = pd.DataFrame({
  'name':df['user1'], 
  'email':df['email1'], 
  'updatedAt': datetime.now().isoformat(),
  'teamId': range(42, 81)
})
user1.head()
```

```python
user2 = pd.DataFrame({
  'name':df['user2'], 
  'email':df['email2'], 
  'updatedAt': datetime.now().isoformat(),
  'teamId': range(42, 81)
})
user2.head()
```

### Save to CSV

```python
teams.to_csv('teams.csv', index=False)
user1.to_csv('user1.csv', index=False)
user2.to_csv('user2.csv', index=False)
```

### Merging user1 and user2 to users.csv

```python

users = pd.concat([user1, user2], ignore_index=True)
users.to_csv('users.csv', index=False)
```
